```python
from yahoo_fin import stock_info as si
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = si.get_data('BTC-USD',start_date='2022-01-01',end_date='2024-12-31')
data
```

```python
dates = np.array(data.index,dtype='datetime64[D]')
dates
```

```python
closing_prices = np.array(data['close'])
closing_prices
```

```python
plt.figure(figsize=(12, 6))
plt.plot(dates, closing_prices, label="Precio de Cierre", color="blue")
plt.title("Tendencia del Precio de Bitcoin", fontsize=14)
plt.xlabel("Fecha", fontsize=12)
plt.ylabel("Precio de Cierre (USD)", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
