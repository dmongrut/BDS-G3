```python
import pandas as pd
df_books = pd.read_csv('https://github.com/cesarmayta/datasets/raw/refs/heads/main/libros.csv')
df_books.head(2)
```

```python
df_books.groupby('Author').sum()
```

```python
df_books.groupby('Author').sum().reset_index()
```

```python
df_books.groupby('Author').agg(['min','max']).reset_index()
```

```python
df_books.groupby('Author').agg({'Reviews':['min','max'],'User Rating':'sum'}).reset_index()
```

```python
df_books.groupby(['Year','Genre']).count().reset_index()
```
