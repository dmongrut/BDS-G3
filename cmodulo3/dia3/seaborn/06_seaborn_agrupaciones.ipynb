```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')
tips.head(2)
```

# JOINTPLOT

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='sex',kind='hist')
plt.show()
```

# PAIRPLOT

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm',corner=True)
plt.show()
```
