```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic = titanic.dropna(subset=["class","alive"])
```

```python
titanic
```

```python
sobrevivientes_por_clase = (
    titanic[titanic['alive'] == 'yes']
    .groupby('class')
    .size()
    .reset_index(name='sobrevivientes')
)
sobrevivientes_por_clase
```

```python
sns.barplot(x='class',y='sobrevivientes',data=sobrevivientes_por_clase,palette='pastel')
```
