```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# CARGAMOS DATASET DE EJEMPLO

```python
tips = sns.load_dataset('tips')
tips
```

# CREAMOS GRAFICO DISPLOT

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

# AGRUPAMIENTO DE DATOS

```python
sns.displot(data=tips,x='total_bill',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=True,palette='dark')
plt.show()
```

# AGREGAMOS LEYENDAS AL GRAFICO

```python
sns.displot(data=tips,x='total_bill',hue='smoker',kind='kde',legend=True,palette='dark')
plt.title('cuenta por fumadores y no fumadores')
plt.xlabel('cuenta')
plt.ylabel('Densidad')
plt.show()
```
