# APLICACIÓN DE MINIMOS CUADRADOS ORDINARIOS CON [STATSMODELS.API.OLS](https://www.statsmodels.org/dev/generated/statsmodels.formula.api.ols.html)

### ¿Qué es ols en statsmodels.formula.api?

ols es la abreviatura de Ordinary Least Squares (Mínimos Cuadrados Ordinarios), un método estadístico para encontrar la mejor línea de ajuste en un modelo de regresión lineal.

Se encuentra en el módulo statsmodels.formula.api y permite definir la relación entre una variable dependiente (respuesta) y una o más variables independientes (predictoras) usando una notación de fórmula.

### ¿Para qué sirve?
Se usa para:

*   Ajustar modelos de regresión lineal de forma sencilla con fórmulas.
*   Analizar relaciones entre variables.
* Obtener coeficientes de regresión, intervalos de confianza y métricas estadísticas.
* Realizar pruebas de hipótesis sobre los coeficientes.

### EJEMPLO BASICO

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
data = pd.DataFrame({
    'peso': [50,60,70,80,90],
    'altura':[1.50,1.60,1.70,1.80,1.90]
})
data
```

### GRAFICAMOS LA RELACIÓN ENTRE X y Y

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['altura'],y=data['peso'],label='Datos Reales',color='blue')
```

# APLICAMOS EL MODELO DE REGRESIÓN LINEAL CON AL AJUSTE DE MINIMOS CUADRADOS ORDINARIOS

```python
modelo = smf.ols(formula='peso ~ altura',data=data).fit()
data['prediccion'] = modelo.predict(data['altura'])
data
```

## GRAFICAMOS LA PREDICCIÓN

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['altura'],y=data['peso'],label='Datos Reales',color='blue')
sns.lineplot(x=data['altura'],y=data['prediccion'],label='Regresión',color='red')
plt.xlabel('Altura (m)')
plt.ylabel('Peso (kg)')
plt.title('Regresión Lineal: Peso vs. Altura')
plt.legend()
plt.show()
```
