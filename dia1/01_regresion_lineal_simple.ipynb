### USO DE SCIPY

Para probar tu regresi贸n lineal con valores que no est谩n en tu conjunto de datos (es decir, para hacer predicciones), puedes utilizar la ecuaci贸n de la l铆nea de regresi贸n que has obtenido de scipy.stats.linregress. La ecuaci贸n es de la forma:


=


+



donde:


1. m es la pendiente (slope)
2. b es el intercepto (intercept)
3. x es el valor nuevo para el cual quieres predecir

```python
import scipy.stats

# datos de ejemplo
x = np.array([0,1,2,3,4])
y = np.array([2,2.8,3.6,4.4,5.2])
print(x,y)
```

### GRAFICAMOS LA RELACIN ENTRE X y Y

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='Datos Observados')
```

### REALIZAMOS LA REGRESIN LINEAL SIMPLE

```python
result = scipy.stats.linregress(x,y)
result
```

### MOSTRAMOS RESULTADOS DE LA REGRESIN LINEAL

```python
print(f'Pendiente(slope): {result.slope}')
print(f'Intercepto(intercept): {result.intercept}')
print(f'Coeficiente de Correlaci贸n(rvalue) {result.rvalue}')
print(f'Valor p(pvalue) : {result.pvalue}')
print(f'Error st谩ndar de la pendiente (stderr) : {result.stderr}')
```

```python
y_pred = result.slope * x + result.intercept
y_pred
```

### GRAFICAMOS LA REGRESIN LINEAL

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='Datos Observados')
plt.plot(x,y_pred,color='red',label='L铆nea de regresi贸n')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresi贸n lineal con scipy')
plt.legend()
plt.show()
```

### PROBAMOS NUESTRA PREDICCIN

```python
x_new = 5
y_new_pred = result.slope * x_new + result.intercept
print(f'predicci贸n para x = {x_new}: y = {y_new_pred}')
```

### GRAFICAMOS EL VALOR DE PREDICCIN

```python

m = result.slope
b = result.intercept
y_pred_x = m * x + b
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='datos observador')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y_pred_x,color='red',label='Linea de regresi贸n')
plt.scatter(x_new,y_new_pred,color='green',marker='o',s=100,label=f'predicci贸n para x={x_new}')

plt.legend()
plt.show()
```
