```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

# PRIMER PLOT CON MATPLOTLIB

```python
plt.plot(x,y)
plt.show()
```

# CAMBIAR EL COLOR DEL GRAFICO

```python
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'ro')
plt.show()
```

```python
plt.plot(x,y,'ro-')
plt.show()
```

```python
plt.plot(x,y,'rD:')
plt.show()
```

# TIPOS DE GRAFICOS

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```
